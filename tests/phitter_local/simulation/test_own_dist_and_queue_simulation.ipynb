```python
import sys


sys.path.insert(0, "../../../")



import phitter
```

```python
parameters = {0: 0.5, 1: 0.3, 2: 0.2}
simulation = phitter.simulation.QueueingSimulation(
    a="exponential",
    a_parameters={"lambda": 5},
    s="exponential",
    s_parameters={"lambda": 20},
    c=3,
    d="PBS",
    pbs_distribution="own_distribution",
    pbs_parameters=parameters,
)
simulation.run(2000)
```

```python
simulation.number_probability_summary()
```

```python
simulation = phitter.simulation.QueueingSimulation(
    a="exponential",
    a_parameters={"lambda": 5},
    s="exponential",
    s_parameters={"lambda": 20},
    c=3,
    d="FIFO",
)



probabilities_confidence_interval, metrics_confidence_interval = (
    simulation.confidence_interval_metrics(2000, replications=10)
)
```
