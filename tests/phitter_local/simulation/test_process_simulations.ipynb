```python
import sys



sys.path.insert(0, "../../../")


import phitter
```

```python
simulation = phitter.simulation.ProcessSimulation()
simulation.add_process(
    prob_distribution="normal",
    parameters={"mu": 5, "sigma": 2},
    process_id="first",
    new_branch=True,
    number_of_products=10,
)
simulation.add_process(
    prob_distribution="exponential",
    parameters={"lambda": 4},
    process_id="second",
    previous_ids=["first"],
)
simulation.add_process(
    prob_distribution="exponential",
    parameters={"lambda": 4},
    process_id="ni_idea",
    previous_ids=["first"],
)
simulation.add_process(
    prob_distribution="exponential",
    parameters={"lambda": 4},
    process_id="ni_idea_2",
    previous_ids=["first"],
    number_of_products=15,
    number_of_servers=4,
)

simulation.add_process(
    prob_distribution="gamma",
    parameters={"alpha": 15, "beta": 3},
    process_id="third",
    new_branch=True,
)
# simulation.add_process("exponential", {"lambda": 4.3}, "nn", previous_ids=["third"])
simulation.add_process(
    prob_distribution="beta",
    parameters={"alpha": 1, "beta": 1, "A": 2, "B": 3},
    process_id="fourth",
    previous_ids=["second", "third"],
    number_of_products=2,
)
simulation.add_process(
    prob_distribution="exponential",
    parameters={"lambda": 4.3},
    process_id="nn",
    previous_ids=["third"],
)

simulation.add_process(
    prob_distribution="exponential",
    parameters={"lambda": 4.3},
    process_id="fifth",
    new_branch=True,
)
simulation.add_process(
    prob_distribution="exponential",
    parameters={"lambda": 4.3},
    process_id="sixth",
    previous_ids=["fourth", "ni_idea", "nn"],
)
simulation.add_process(
    prob_distribution="beta",
    parameters={"alpha": 1, "beta": 1, "A": 2, "B": 3},
    process_id="seventh",
    previous_ids=["fifth", "sixth", "ni_idea_2"],
)
```

```python
simulation.run(10)
```

```python
simulation.simulation_metrics()
```

```python
simulation.run_confidence_interval(number_of_simulations=100)
```
