```python
import sys
sys.path.append("../../../../")
import phitter
```

```python
def get_data(path: str) -> list[float | int]:
    sample_distribution_file = open(path, "r")
    data = [float(x.replace(",", ".")) for x in sample_distribution_file.read().splitlines()]
    sample_distribution_file.close()
    return data
```

```python
path = "../../../../datasets_test/discrete/sample_geometric.txt"
data = get_data(path)
```

```python
data = phitter.discrete.BINOMIAL(init_parameters_examples=True).sample(10000)
```

```python
phi = phitter.PHITTER(data, fit_type="discrete")
phi.fit(n_workers=1)
```

```python
phi.plot_histogram()
```

```python
phi.plot_histogram_distributions()
```

```python
phi.plot_distribution("hypergeometric")
```
