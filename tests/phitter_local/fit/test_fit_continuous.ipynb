```python
sys.path.append("../../../")
import phitter
```

```python
def get_data(path: str) -> list[float | int]:
    sample_distribution_file = open(path, "r")
    data = [float(x.replace(",", ".")) for x in sample_distribution_file.read().splitlines()]
    sample_distribution_file.close()
    return data
```

```python
path = "../../../phitter/continuous/continuous_distributions_sample/sample_beta.txt"
data = get_data(path)
```

```python
phi = phitter.Phitter(data=data)
phi.fit(n_workers=1)
```
