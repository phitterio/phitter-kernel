```python
import phitter
```

```python
df = pandas.read_csv("https://raw.githubusercontent.com/lux-org/lux-datasets/master/data/college.csv")
df.head()
```

```python
data = df["MedianFamilyIncome"].to_numpy()
```

```python
phi = phitter.Phitter(data=data=data)
phi.fit(n_workers=2)
```

```python
phi.best_distribution
```

```python
phi.df_sorted_distributions_sse.head(10)
```

```python
phi.plot_histogram()
```

```python
phi.plot_histogram_distributions()
```

```python
phi.plot_distribution("johnson_sb")
```

```python
phi.plot_ecdf()
```

```python
phi.plot_ecdf_distribution("johnson_sb")
```
