```python
import phitter
```

```python
df_estimation_time = pandas.DataFrame()
```

```python
SAMPLE_SIZES = [1000, 10000, 100000, 500000, 1000000, 10000000]
```

```python
for id_distribution, distribution_class in phitter.continuous.CONTINUOUS_DISTRIBUTIONS.items():
    for size in SAMPLE_SIZES:
        sample = distribution_class(init_parameters_examples=True).sample(size)
        continuous_measures = phitter.continuous.ContinuousMeasures(sample)
        ti = time.time()
        distribution = distribution_class(continuous_measures=continuous_measures)
        tf = time.time() - ti
        df_estimation_time.loc[id_distribution, size] = tf
```
