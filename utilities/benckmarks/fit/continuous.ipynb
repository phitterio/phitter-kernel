```python
import phitter
```

```python
df_fit_time = pandas.DataFrame()
```

```python
df_fit_time
```

```python
IDS_DISTRIBUTION_FIT = ["beta", "fatigue_life", "triangular", "exponential_3p", "power_funtion", "normal", "f_4p", "johnson_sb"]
SAMPLE_SIZES = [100, 1000, 10000, 100000, 1000000]
WORKERS = [1, 2, 6, 10, 20]
```

```python
for id_distribution in IDS_DISTRIBUTION_FIT:
    for n_workers in WORKERS:
        for sample_size in SAMPLE_SIZES:
            distribution_class = phitter.continuous.ALL_CONTINUOUS_DISTRIBUTIONS["beta"]
            data = distribution_class(init_parameters_examples=True).sample(sample_size)
            ti = time.time()
            phitter_cont = phitter.PHITTER(data=data)
            phitter_cont.fit(n_workers=n_workers)
            tf = time.time() - ti
            if n_workers not in df_fit_time.columns and sample_size not in df_fit_time.index:
                df_fit_time.loc[sample_size, n_workers] = tf
            else:
                df_fit_time.loc[sample_size, n_workers] += tf
```

```python
df_fit_time = df_fit_time / len(IDS_DISTRIBUTION_FIT)
```
